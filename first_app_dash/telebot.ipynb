```python
#config.py
import os
from dotenv import load_dotenv

load_dotenv()

class Config:
    # Telegram Bot Token
    BOT_TOKEN = os.getenv('BOT_TOKEN')
    
    # Staff Permissions
    ADMIN_IDS = [int(x) for x in os.getenv('ADMIN_IDS', '').split(',') if x]
    OPERATOR_IDS = [int(x) for x in os.getenv('OPERATOR_IDS', '').split(',') if x]
    
    # Broadcast Channel
    BROADCAST_CHANNEL = os.getenv('BROADCAST_CHANNEL', '@ExtensoVillageAnnouncements')
    
    # Database
    DATABASE_URL = 'sqlite:///extenso_village.db'
    
    # File paths
    UPLOAD_FOLDER = 'receipts'
    DATA_FOLDER = 'data'

class Permissions:
    @staticmethod
    def is_admin(user_id):
        return user_id in Config.ADMIN_IDS
    
    @staticmethod
    def is_operator(user_id):
        return user_id in Config.OPERATOR_IDS or user_id in Config.ADMIN_IDS
    
    @staticmethod
    def require_admin(func):
        def wrapper(update, context):
            if not Permissions.is_admin(update.effective_user.id):
                update.message.reply_text("âŒ Administrator access required.")
                return
            return func(update, context)
        return wrapper
    
    @staticmethod
    def require_operator(func):
        def wrapper(update, context):
            if not Permissions.is_operator(update.effective_user.id):
                update.message.reply_text("âŒ Operator access required.")
                return
            return func(update, context)
        return wrapper
```

```python
# database.py

import sqlite3
import datetime
import os
from config import Config

class Database:
    def __init__(self):
        self.db_path = 'extenso_village.db'
        self.init_database()
        
    def get_connection(self):
        # Using check_same_thread=False for flexibility in bot threads
        conn = sqlite3.connect(self.db_path, check_same_thread=False)
        conn.row_factory = sqlite3.Row
        return conn
    
    def init_database(self):
        os.makedirs(Config.DATA_FOLDER, exist_ok=True)
        os.makedirs(Config.UPLOAD_FOLDER, exist_ok=True)
        
        conn = self.get_connection()
        cursor = conn.cursor()
        
        # Residents table (Master Data)
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS residents (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                unit TEXT UNIQUE NOT NULL,
                owner_name TEXT NOT NULL,
                phone TEXT,
                telegram_id INTEGER UNIQUE,
                telegram_name TEXT,
                current_balance INTEGER DEFAULT 0,
                monthly_charge INTEGER DEFAULT 50000,
                last_payment_date TEXT,
                registration_date TEXT,
                language TEXT DEFAULT 'en',
                is_active BOOLEAN DEFAULT 1,
                created_at TEXT DEFAULT CURRENT_TIMESTAMP
            )
        ''')
        
        # Payments table (Pending/Verified Payments)
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS payments (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                unit TEXT NOT NULL,
                amount INTEGER NOT NULL,
                reference TEXT NOT NULL,
                receipt_filename TEXT,
                submission_date TEXT NOT NULL,
                status TEXT DEFAULT 'pending',
                verified_by TEXT,
                verification_date TEXT,
                telegram_user_id INTEGER,
                language TEXT,
                created_at TEXT DEFAULT CURRENT_TIMESTAMP,
                FOREIGN KEY (unit) REFERENCES residents (unit)
            )
        ''')
        
        # NOTE: Support conversations table creation would go here if needed.
        
        self._init_sample_data(conn)
        conn.commit()
        conn.close()
        print("âœ… Database initialized!")

    def _init_sample_data(self, conn):
        cursor = conn.cursor()
        cursor.execute("SELECT COUNT(*) FROM residents")
        if cursor.fetchone()[0] == 0:
            sample_residents = [
                ('88/01', 'John Smith', '0812345678', 50000), # à¸¿500.00 charge
                ('88/02', 'Sarah Chen', '0898765432', 50000),
                ('88/03', 'Mike Johnson', '0855512345', 75000), # à¸¿750.00 charge
            ]
            
            for unit, name, phone, charge in sample_residents:
                cursor.execute('''
                    INSERT INTO residents (unit, owner_name, phone, monthly_charge, registration_date)
                    VALUES (?, ?, ?, ?, ?)
                ''', (unit, name, phone, charge, datetime.datetime.now().strftime('%Y-%m-%d')))
    
    # --- Resident/User Operations ---
    
    def get_resident_by_telegram(self, telegram_id):
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM residents WHERE telegram_id = ?', (telegram_id,))
        result = cursor.fetchone()
        conn.close()
        return dict(result) if result else None
    
    def get_user_account(self, telegram_id):
        """Fetches account data (used for /balance and /payment init)."""
        resident = self.get_resident_by_telegram(telegram_id)
        if resident:
            return {
                'unit': resident['unit'],
                'balance': resident['current_balance'],
                'monthly_charge': resident['monthly_charge'],
                'language': resident['language']
            }
        return None

    # **CRITICAL for /broadcast and /remind**
    def get_all_registered_residents(self):
        """Fetches all residents who have linked a Telegram ID."""
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('''
            SELECT unit, telegram_id, language, monthly_charge
            FROM residents 
            WHERE telegram_id IS NOT NULL AND is_active = 1
        ''')
        results = [dict(row) for row in cursor.fetchall()]
        conn.close()
        return results

    def get_resident_by_unit(self, unit):
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('SELECT * FROM residents WHERE unit = ?', (unit,))
        result = cursor.fetchone()
        conn.close()
        return dict(result) if result else None
    
    def update_resident_language(self, unit, language):
        """Updates the resident's language preference."""
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('UPDATE residents SET language = ? WHERE unit = ?', (language, unit))
        success = cursor.rowcount > 0
        conn.commit()
        conn.close()
        return success
    
    def register_resident(self, unit, telegram_id, telegram_name, language='en'):
        """Links a Telegram ID to a unit, completing registration."""
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('''
            UPDATE residents 
            SET telegram_id = ?, telegram_name = ?, language = ?
            WHERE unit = ? AND telegram_id IS NULL
        ''', (telegram_id, telegram_name, language, unit))
        success = cursor.rowcount > 0
        conn.commit()
        conn.close()
        return success
    
    # --- Payment Operations ---
    
    def log_new_payment(self, unit, amount, reference, receipt_filename, telegram_id, language):
        """Logs a new pending payment submission (used by app.py)."""
        return self.save_payment(unit, amount, reference, receipt_filename, telegram_id, language)

    def save_payment(self, unit, amount, reference, receipt_filename, telegram_id, language):
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('''
            INSERT INTO payments (unit, amount, reference, receipt_filename, submission_date, telegram_user_id, language)
            VALUES (?, ?, ?, ?, ?, ?, ?)
        ''', (unit, amount, reference, receipt_filename, 
              datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'), 
              telegram_id, language))
        payment_id = cursor.lastrowid
        conn.commit()
        conn.close()
        return payment_id
    
    def get_pending_payments(self):
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('''
            SELECT p.id, p.unit, p.amount, p.reference, p.receipt_filename, r.owner_name, p.language
            FROM payments p 
            JOIN residents r ON p.unit = r.unit 
            WHERE p.status = 'pending'
            ORDER BY p.submission_date
        ''')
        results = [dict(row) for row in cursor.fetchall()]
        conn.close()
        return results
    
    def get_payment_by_id(self, payment_id):
        conn = self.get_connection()
        cursor = conn.cursor()
        cursor.execute('SELECT id, unit, amount, reference, receipt_filename, telegram_user_id, language FROM payments WHERE id = ?', (payment_id,))
        result = cursor.fetchone()
        conn.close()
        return dict(result) if result else None
    
    def update_payment_status(self, payment_id, status, verified_by):
        """Updates payment status and credits/debits the resident's account."""
        conn = self.get_connection()
        cursor = conn.cursor()
        
        cursor.execute('SELECT unit, amount FROM payments WHERE id = ?', (payment_id,))
        payment = cursor.fetchone()
        if not payment:
            conn.close()
            return False
        
        unit, amount = payment['unit'], payment['amount']
        
        cursor.execute('''
            UPDATE payments 
            SET status = ?, verified_by = ?, verification_date = ?
            WHERE id = ?
        ''', (status, verified_by, datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'), payment_id))
        
        if status == 'verified':
            # Add amount to current_balance
            cursor.execute('''
                UPDATE residents 
                SET current_balance = current_balance + ?, last_payment_date = ?
                WHERE unit = ?
            ''', (amount, datetime.datetime.now().strftime('%Y-%m-%d'), unit))
        
        conn.commit()
        conn.close()
        return True

# Initialize the Database instance
db = Database()
```

```python
# staff_commands.py

from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import CallbackContext
from config import Config, Permissions 
from database import db # Use the shared db instance 
from messages import Messages 

class BilingualStaffCommands:
    def get_user_language(self, user_id):
        """
        Get staff member's preferred language based on config lists.
        Defaults to 'th' for Thai operators and 'en' for English staff/admins.
        """
        if user_id in Config.THAI_STAFF_IDS:
            return 'th'
        elif user_id in Config.ADMIN_IDS or user_id in Config.ENGLISH_STAFF_IDS:
            return 'en'
        return 'th'
    
    # ==================== CALLBACK HANDLER ====================

    def handle_payment_callback(self, update: Update, context: CallbackContext):
        """Processes the inline button press (approve, reject, receipt) from pending_payments."""
        query = update.callback_query
        query.answer()
        user_id = query.from_user.id
        lang = self.get_user_language(user_id)
        
        action, payment_id_str = query.data.split('_')
        payment_id = int(payment_id_str)

        if action == 'receipt':
            # ACTION: View Receipt
            payment = db.get_payment_by_id(payment_id)
            
            # Use 'receipt_filename' as this holds the Telegram file_id
            if payment and payment.get('receipt_filename'): 
                try:
                    # Send the photo to the staff member
                    context.bot.send_photo(user_id, payment['receipt_filename'])
                    # Update the original message caption to acknowledge
                    query.edit_message_caption(
                        caption=query.message.caption + f"\n\n({Messages.get_message('receipt_sent_confirmation', lang)})" 
                        # NOTE: Add receipt_sent_confirmation to messages.py
                    )
                except Exception:
                    query.edit_message_text(Messages.get_message('receipt_not_found', lang)) 
            else:
                query.edit_message_text(Messages.get_message('receipt_not_found', lang))
            return

        if action in ['approve', 'reject']:
            # ACTION: Approve or Reject Payment
            status = 'verified' if action == 'approve' else 'rejected'
            
            if db.update_payment_status(payment_id, status, query.from_user.first_name): 
                payment = db.get_payment_by_id(payment_id)
                amount_rm = payment['amount'] / 100
                
                # --- Step 1: Notify Resident ---
                resident_lang = payment.get('language', 'en') 
                message_key = 'payment_verified' if status == 'verified' else 'payment_rejected_resident' 
                
                context.bot.send_message( 
                    chat_id=payment['telegram_user_id'],
                    text=Messages.get_message(message_key, resident_lang, 
                                            amount=amount_rm, 
                                            reference=payment['reference'])
                )
                
                # --- Step 2: Confirm to Operator ---
                confirmation_msg = Messages.get_message(f'payment_{status}', lang, 
                                                        payment_id=payment_id,
                                                        amount=amount_rm)
                # Edit the original alert message to remove buttons and show confirmation
                query.edit_message_text(confirmation_msg) 
            else:
                query.edit_message_text(Messages.get_message('payment_not_found', lang))

    # ==================== OPERATOR COMMANDS ====================
    
    @Permissions.require_operator
    def pending_payments(self, update: Update, context: CallbackContext):
        """Show pending payments - sends one message per payment with inline buttons."""
        user_id = update.effective_user.id
        lang = self.get_user_language(user_id)
        
        pending = db.get_pending_payments()
        
        if not pending:
            update.message.reply_text(Messages.get_message('no_pending', lang))
            return
        
        update.message.reply_text(
            Messages.get_message('pending_summary', lang, count=len(pending)), 
            parse_mode='Markdown'
        )

        for payment in pending:
            amount_rm = payment['amount'] / 100
            
            # The 'unit' column is used for the house ID
            text = Messages.get_message('payment_alert_details', lang, 
                                        payment_id=payment['id'],
                                        unit=payment['unit'],
                                        owner_name=payment['owner_name'],
                                        amount=amount_rm, 
                                        reference=payment['reference'])
            
            # Inline Keyboard for action buttons
            keyboard = [[
                InlineKeyboardButton(Messages.get_message('approve', lang), callback_data=f"approve_{payment['id']}"),
                InlineKeyboardButton(Messages.get_message('reject', lang), callback_data=f"reject_{payment['id']}")
            ], [
                InlineKeyboardButton(Messages.get_message('view_receipt', lang), callback_data=f"receipt_{payment['id']}")
            ]]
            reply_markup = InlineKeyboardMarkup(keyboard)
            
            # Send the payment details as a message with buttons attached
            update.message.reply_text(text, reply_markup=reply_markup, parse_mode='Markdown')
            
    @Permissions.require_operator
    def operator_dashboard(self, update: Update, context: CallbackContext):
        """Operator dashboard in their language, showing pending counts."""
        user_id = update.effective_user.id
        lang = self.get_user_language(user_id)
        pending = db.get_pending_payments()
        
        text = Messages.get_message('operator_dashboard', lang, pending_count=len(pending))
        update.message.reply_text(text, parse_mode='Markdown')
    
    # ==================== ADMIN COMMANDS ====================
    
    @Permissions.require_admin
    def broadcast_message(self, update: Update, context: CallbackContext):
        """Admin broadcast - sends announcement in resident's language."""
        if not context.args:
            # Show usage in admin's language (English)
            update.message.reply_text(Messages.get_message('broadcast_usage', 'en'))
            return
        
        message = ' '.join(context.args)
        residents = db.get_all_registered_residents()
        
        success_count = 0
        for resident in residents:
            try:
                lang = resident.get('language', 'en')
                context.bot.send_message(
                    chat_id=resident['telegram_id'],
                    text=Messages.get_message('broadcast_announcement', lang, message=message),
                    parse_mode='Markdown'
                )
                success_count += 1
            except Exception as e:
                print(f"Failed to send broadcast to {resident['unit']}: {e}")
        
        # Confirm to admin in English
        update.message.reply_text(Messages.get_message('broadcast_sent', 'en', count=success_count))
    
    @Permissions.require_admin
    def admin_dashboard(self, update: Update, context: CallbackContext):
        """Admin dashboard - always in English."""
        residents = db.get_all_registered_residents()
        pending = db.get_pending_payments()
        
        text = Messages.get_message('admin_dashboard', 'en',
                                  resident_count=len(residents),
                                  pending_count=len(pending))
        
        # Add admin commands in English
        text += f"\n**Admin Commands:**"
        text += f"\n/broadcast [message] - Send announcement"
        text += f"\n/remind - Send payment reminders" 
        
        update.message.reply_text(text, parse_mode='Markdown')
    
    @Permissions.require_admin
    def payment_reminder(self, update: Update, context: CallbackContext):
        """Send payment reminders in resident's language to all registered residents."""
        residents = db.get_all_registered_residents()
        
        success_count = 0
        for resident in residents:
            try:
                lang = resident.get('language', 'en')
                # Assumes monthly_charge is available in the resident dict from the DB query
                context.bot.send_message(
                    chat_id=resident['telegram_id'],
                    text=Messages.get_message('payment_reminder', lang, amount=resident['monthly_charge']/100),
                    parse_mode='Markdown'
                )
                success_count += 1
            except Exception as e:
                print(f"Failed to send reminder to {resident['unit']}: {e}")
        
        # Confirm to admin in English
        update.message.reply_text(Messages.get_message('reminders_sent', 'en', count=success_count))
```

```python
#config.py
import os
from dotenv import load_dotenv

load_dotenv()

class Config:
    BOT_TOKEN = os.getenv('BOT_TOKEN')
    
    # Staff permissions with language preference
    ADMIN_IDS = [int(x) for x in os.getenv('ADMIN_IDS', '').split(',') if x]
    OPERATOR_IDS = [int(x) for x in os.getenv('OPERATOR_IDS', '').split(',') if x]
    
    # Language preferences
    ENGLISH_STAFF_IDS = [int(x) for x in os.getenv('ENGLISH_STAFF_IDS', '').split(',') if x]
    THAI_STAFF_IDS = [int(x) for x in os.getenv('THAI_STAFF_IDS', '').split(',') if x]
    
    BROADCAST_CHANNEL = os.getenv('BROADCAST_CHANNEL', '@ExtensoVillageAnnouncements')
    DATABASE_URL = 'sqlite:///extenso_village.db'
    UPLOAD_FOLDER = 'receipts'
    DATA_FOLDER = 'data'

class Permissions:
    @staticmethod
    def is_admin(user_id):
        return user_id in Config.ADMIN_IDS
    
    @staticmethod
    def is_operator(user_id):
        return user_id in Config.OPERATOR_IDS or user_id in Config.ADMIN_IDS
    
    @staticmethod
    def require_admin(func):
        def wrapper(update, context):
            if not Permissions.is_admin(update.effective_user.id):
                lang = 'th' if update.effective_user.id in Config.THAI_STAFF_IDS else 'en'
                from messages import Messages
                update.message.reply_text(Messages.get_message('admin_access_required', lang))
                return
            return func(update, context)
        return wrapper
    
    @staticmethod
    def require_operator(func):
        def wrapper(update, context):
            if not Permissions.is_operator(update.effective_user.id):
                lang = 'th' if update.effective_user.id in Config.THAI_STAFF_IDS else 'en'
                from messages import Messages
                update.message.reply_text(Messages.get_message('staff_access_required', lang))
                return
            return func(update, context)
        return wrapper
```

```python
# app.py

import logging
import re
from telegram import Update, InlineKeyboardButton, InlineKeyboardMarkup
from telegram.ext import (
    Updater, CommandHandler, CallbackQueryHandler, MessageHandler, Filters, ConversationHandler, CallbackContext
)
from config import Config, Permissions 
from messages import Messages 
from database import db 
from staff_commands import BilingualStaffCommands 

# Enable logging
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO) 

# --- Conversation States ---
UNIT_ENTRY, PIN_ENTRY = range(2) # For /register
RECEIPT_UPLOAD, DETAILS_ENTRY = range(2, 4) # For /payment

class BilingualExtensoBot:
    def get_user_language(self, user_id):
        """Get user's language preference (resident or staff)."""
        resident = db.get_resident_by_telegram(user_id) 
        if resident:
            return resident.get('language', 'en') 
        
        # Staff language preferences
        if user_id in Config.THAI_STAFF_IDS:
            return 'th'
        elif user_id in Config.ENGLISH_STAFF_IDS or user_id in Config.ADMIN_IDS:
            return 'en'
        else:
            return 'en' # Default language for non-registered users

    # --- GENERAL COMMANDS & CALLBACKS ---
    
    def start(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        user = update.effective_user
        lang = self.get_user_language(user.id)
        
        if Permissions.is_operator(user.id):
            message_key = 'welcome_admin' if Permissions.is_admin(user.id) else 'welcome_operator'
            update.message.reply_text(Messages.get_message(message_key, lang))
            return
        
        # Residents choose language
        keyboard = [
            [InlineKeyboardButton("ðŸ‡¬ðŸ‡§ English", callback_data='lang_en')],
            [InlineKeyboardButton("ðŸ‡¹ðŸ‡­ à¹„à¸—à¸¢", callback_data='lang_th')]
        ]
        reply_markup = InlineKeyboardMarkup(keyboard)
        
        update.message.reply_text(
            Messages.get_message('welcome', 'en'),
            reply_markup=reply_markup,
            parse_mode='Markdown'
        )
    
    def button_handler(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        query = update.callback_query
        query.answer()
        user = query.from_user
        
        if query.data.startswith('lang_'):
            language = query.data.split('_')[1]
            
            # Update user's language preference in the DB
            resident = db.get_resident_by_telegram(user.id)
            if resident:
                 db.update_resident_language(resident['unit'], language)
            
            # For non-registered users, we can just display the prompt
            query.edit_message_text(
                Messages.get_message('language_set', language) + "\n\n" +
                Messages.get_message('register_prompt', language),
                parse_mode='Markdown'
            )
    
    # --- REGISTRATION CONVERSATION HANDLER (NEW) ---
    
    def start_registration(self, update: Update, context: CallbackContext):
        user = update.effective_user
        user_id = user.id
        lang = self.get_user_language(user_id)

        if db.get_resident_by_telegram(user_id):
            update.message.reply_text(Messages.get_message('ALREADY_REGISTERED', lang, unit=db.get_resident_by_telegram(user_id)['unit']))
            return ConversationHandler.END

        # Check if unit is provided immediately (e.g., /register 88/01)
        if context.args and len(context.args) == 1:
            unit = context.args[0].upper().replace('/', '-') # Normalize unit format
            context.user_data['reg_unit'] = unit
            context.user_data['reg_lang'] = lang # Store language for consistent replies
            return self.receive_unit(update, context, initial_unit=unit)
        
        context.user_data['reg_lang'] = lang 
        update.message.reply_text(Messages.get_message('PROMPT_HOUSE_ID', lang))
        return UNIT_ENTRY

    def receive_unit(self, update: Update, context: CallbackContext, initial_unit=None):
        user = update.effective_user
        lang = context.user_data.get('reg_lang', self.get_user_language(user.id))
        
        unit = initial_unit if initial_unit else update.message.text.strip().upper().replace('/', '-')

        resident = db.get_resident_by_unit(unit)

        if not resident:
            update.message.reply_text(Messages.get_message('unit_not_found', lang, unit=unit))
            return UNIT_ENTRY

        if resident['telegram_id'] is not None:
            update.message.reply_text(Messages.get_message('UNIT_ALREADY_LINKED', lang, unit=unit))
            return ConversationHandler.END

        context.user_data['reg_unit'] = unit
        context.user_data['owner_name'] = resident['owner_name']
        
        update.message.reply_text(Messages.get_message('PROMPT_PASSWORD', lang))
        return PIN_ENTRY

    def receive_pin(self, update: Update, context: CallbackContext):
        user = update.effective_user
        lang = context.user_data.pop('reg_lang', self.get_user_language(user.id))
        unit = context.user_data.pop('reg_unit', None)
        owner_name = context.user_data.pop('owner_name', 'Resident')
        
        pin = update.message.text.strip()
        
        if not unit:
            update.message.reply_text(Messages.get_message('error_session_expired', lang))
            return ConversationHandler.END

        # --- PIN VALIDATION (Placeholder: Assume all units use '1234') ---
        # NOTE: In a real system, you'd check a PIN column in the database here.
        if pin != '1234': 
            update.message.reply_text(Messages.get_message('INVALID_CREDENTIALS', lang))
            return PIN_ENTRY
        # -----------------------------------------------------------------

        if db.register_resident(unit, user.id, user.username or user.first_name, lang):
            update.message.reply_text(
                Messages.get_message('register_success', lang, 
                                     unit=unit, 
                                     name=owner_name, 
                                     user_name=user.username or user.first_name),
                parse_mode='Markdown'
            )
        else:
            # Fallback for race condition/unexpected error
            update.message.reply_text(Messages.get_message('UNIT_ALREADY_LINKED', lang, unit=unit))

        return ConversationHandler.END

    # --- BALANCE CHECK ---
    
    def handle_balance(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        user_id = update.effective_user.id 
        account_data = db.get_user_account(user_id) 
        lang = self.get_user_language(user_id) 

        if not account_data: 
            update.message.reply_text(Messages.get_message('not_registered', lang))
            return

        unit = account_data['unit']
        balance = account_data['balance'] 
        monthly_charge = account_data['monthly_charge']

        balance_baht = balance / 100
        monthly_charge_baht = monthly_charge / 100

        text = Messages.get_message('balance_info', lang, 
                                    unit=unit, 
                                    balance=balance_baht,
                                    monthly_charge=monthly_charge_baht)

        if balance > 0: 
            text += "\n\n" + Messages.get_message('balance_credit', lang, amount=balance_baht)
        elif balance < 0: 
            text += "\n\n" + Messages.get_message('balance_outstanding', lang, amount=abs(balance_baht))

        update.message.reply_text(text, parse_mode='Markdown')

    # --- PAYMENT CONVERSATION ---
    
    def report_payment_start(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        user_id = update.effective_user.id
        account_data = db.get_user_account(user_id)
        lang = self.get_user_language(user_id)

        if not account_data:
            update.message.reply_text(Messages.get_message('not_registered', lang))
            return ConversationHandler.END # Added END here

        context.user_data['payment_unit'] = account_data['unit']
        context.user_data['payment_lang'] = lang 
        
        text = Messages.get_message('payment_instructions', lang, 
                                    unit=account_data['unit'],
                                    balance=account_data['balance'] / 100,
                                    monthly_charge=account_data['monthly_charge'] / 100)
        
        update.message.reply_text(text, parse_mode='Markdown')
        return RECEIPT_UPLOAD

    def receive_receipt_photo(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        user_id = update.effective_user.id
        lang = context.user_data.get('payment_lang', self.get_user_language(user_id))
        
        if not update.message.photo:
            update.message.reply_text(Messages.get_message('receipt_photo_required', lang))
            return RECEIPT_UPLOAD

        file_id = update.message.photo[-1].file_id 
        context.user_data['receipt_file_id'] = file_id 

        update.message.reply_text(Messages.get_message('receipt_received', lang))
        return DETAILS_ENTRY

    def receive_payment_details(self, update: Update, context: CallbackContext):
        # ... (unchanged) ...
        user_id = update.effective_user.id
        lang = context.user_data.pop('payment_lang', self.get_user_language(user_id)) 
        unit = context.user_data.pop('payment_unit', None) 
        receipt_file_id = context.user_data.pop('receipt_file_id', None) 
        
        if not unit or not receipt_file_id: 
            update.message.reply_text(Messages.get_message('error_session_expired', lang))
            return ConversationHandler.END

        text = update.message.text.strip().replace(',', '') 
        
        try:
            amount_match = re.search(r'(?i)amount\s*:?\s*(\d+(\.\d+)?)', text)
            ref_match = re.search(r'(?i)reference\s*:?\s*([a-zA-Z0-9]+)', text)
            
            if not amount_match or not ref_match:
                 raise ValueError("Missing amount or reference.")

            amount_baht = float(amount_match.group(1))
            reference = ref_match.group(1)
            amount_cents = int(round(amount_baht * 100))
        except Exception:
            update.message.reply_text(Messages.get_message('details_format_error', lang))
            return DETAILS_ENTRY

        payment_id = db.log_new_payment(unit, amount_cents, reference, receipt_file_id, user_id, lang) 

        update.message.reply_text(
            Messages.get_message('payment_submitted', lang, 
                                 unit=unit, 
                                 amount=amount_baht, 
                                 reference=reference),
            parse_mode='Markdown'
        )
        
        # TODO: Notify staff here

        return ConversationHandler.END

    def cancel_conversation(self, update, context):
        # ... (unchanged) ...
        user_id = update.effective_user.id
        lang = self.get_user_language(user_id)
        # Clear user data for the current conversation flow
        context.user_data.clear() 
        update.message.reply_text(Messages.get_message('registration_cancelled', lang))
        return ConversationHandler.END

# --- MAIN EXECUTION ---

def main():
    
    # Instantiate the necessary classes
    staff_commands = BilingualStaffCommands()
    main_bot = BilingualExtensoBot()

    updater = Updater(Config.BOT_TOKEN, use_context=True) 
    dispatcher = updater.dispatcher 
    
    # A. COMMAND HANDLERS
    dispatcher.add_handler(CommandHandler("start", main_bot.start))
    dispatcher.add_handler(CommandHandler("balance", main_bot.handle_balance))
    
    # Staff Commands
    dispatcher.add_handler(CommandHandler("dashboard", staff_commands.operator_dashboard))
    dispatcher.add_handler(CommandHandler("pending", staff_commands.pending_payments))
    # Admin Commands
    dispatcher.add_handler(CommandHandler("broadcast", staff_commands.broadcast_message))
    dispatcher.add_handler(CommandHandler("remind", staff_commands.payment_reminder))

    # B. REGISTRATION CONVERSATION HANDLER (NEW)
    registration_handler = ConversationHandler(
        entry_points=[CommandHandler('register', main_bot.start_registration)],
        states={
            UNIT_ENTRY: [MessageHandler(Filters.text & ~Filters.command, main_bot.receive_unit)],
            PIN_ENTRY: [MessageHandler(Filters.text & ~Filters.command, main_bot.receive_pin)]
        },
        fallbacks=[CommandHandler('cancel', main_bot.cancel_conversation)],
        allow_reentry=True
    )
    dispatcher.add_handler(registration_handler) 

    # C. PAYMENT CONVERSATION HANDLER
    payment_handler = ConversationHandler(
        entry_points=[CommandHandler('payment', main_bot.report_payment_start)],
        states={
            RECEIPT_UPLOAD: [MessageHandler(Filters.photo & ~Filters.command, main_bot.receive_receipt_photo)],
            DETAILS_ENTRY: [MessageHandler(Filters.text & ~Filters.command, main_bot.receive_payment_details)]
        },
        fallbacks=[CommandHandler('cancel', main_bot.cancel_conversation)],
        allow_reentry=True
    )
    dispatcher.add_handler(payment_handler) 

    # D. CALLBACK QUERY HANDLERS
    dispatcher.add_handler(CallbackQueryHandler(staff_commands.handle_payment_callback, pattern='^(approve|reject|receipt)_'))
    dispatcher.add_handler(CallbackQueryHandler(main_bot.button_handler))

    # E. Start the Bot
    print("Starting ExtensoBot polling...")
    updater.start_polling()
    updater.idle()

if __name__ == '__main__':
    # NOTE: Config class and db instance must be initialized before main()
    main()
```
